```python
## load in train/test data.
train_file = "080327_080327.txt_train.txt"
test_file = "080327_080327.txt_test.txt"
import os
def load_data(file, verbose=False):
    file = os.path.join(os.getcwd(), "data", file)
    if verbose:
        print(f"loading file: {file}")
    d = {}
    with open(file, "r") as f:
        edges = f.readlines()
        for edge in edges:
            [a, b] = edge.split()
            if d.get(a) is None:
                d[a] = {b:{}}
            else:
                d[a][b] = {}
            if d.get(b) is None:
                d[b] = {a:{}}
            else:
                d[b][a] = {}
    return nx.Graph(d)
trainG = load_data(train_file)
testG = load_data(test_file)

adj_train = nx.to_scipy_sparse_matrix(trainG)
adj_test = nx.to_scipy_sparse_matrix(testG)
```
